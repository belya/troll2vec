```python
import json
import pandas as pd
```

```python
pikabu_comments_df = pd.read_csv("pikabu/prepared_dataset.csv")
```

```python
dvach_comments_df = pd.read_csv("2ch/posts_2ch_lemmatized.csv")
```

# Toxic texts

2ch texts:
- banned
- with big number of dislikes and big number of answers

```python
dvach_comments_df[dvach_comments_df["banned"] == 1]["preprocessed_text"].sample(10).tolist()
```

```python
dvach_comments_df[dvach_comments_df["toxic"]]["preprocessed_text"].sample(10).tolist()
```

```python
dvach_comments_df[dvach_comments_df["banned"] == 1]["preprocessed_text"].sample(10).tolist()
```

```python
toxic_comments = dvach_comments_df[dvach_comments_df["toxic"] | (dvach_comments_df["banned"] == 1)]["preprocessed_text"]
```

```python
toxic_comments_filtered = [text for text in toxic_comments.astype(str).tolist() if len(text) > 10]
```

# Non toxic texts

Pikabu texts with big number of likes and no dislikes

```python
pikabu_comments_df = pikabu_comments_df.set_index("id")
```

```python
answers = pikabu_comments_df.groupby("pid").size()
```

```python
pikabu_comments_df["answers"] = answers
```

```python
pikabu_comments_df["answers"] = pikabu_comments_df["answers"].fillna(0)
```

```python
non_toxic_comments = pikabu_comments_df[(pikabu_comments_df["dislikes"] == 0) & (pikabu_comments_df["likes"] > 1) & (pikabu_comments_df["answers"] == 0)]["normalized_post"]
```

```python
non_toxic_comments_filtered = [text for text in non_toxic_comments.astype(str).tolist() if len(text) > 10]
```

# Save

```python
with open('toxicity.pos', "w") as file:
    for comment in toxic_comments_filtered:
        file.write(comment + '\n')
```

```python
with open('toxicity.neg', "w") as file:
    for comment in non_toxic_comments_filtered:
        file.write(comment + '\n')
```
